## Data Analysis & Data Visualisation

```python
# Medal count by country, for the top 10 countries
medals_by_country = df[~df['Medal'].isin(['No Medal'])].groupby('region')['Medal'].count().sort_values(ascending=False)[:10]

print('Table 1: Total Medals by Country (Top 10):')
print(medals_by_country)
```

```python
# Visualise Table 1 as a vertical bar plot 

plt.figure(figsize=(10, 4))
plt.bar(medals_by_country.index,medals_by_country)
medals_by_country.plot(kind='bar', stacked=True)
plt.xlabel('Country')
plt.ylabel('Medal Count')
plt.title('Figure 1: Total Medal Count by Country (Top 10)')
plt.xticks(rotation=00)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('Figure 1: Total Medal Count by Country (Top 10).png', dpi=300)
plt.show()
```

Figure 1 illustrates the top 10 countries with the largest olympic medal count, between the 1896-2016 games. It is clearly evident that USA is the outright leader, with over 5000 medals in total (5637 medals, as displayed in Table 1). The USA's total medal count is roughly 35% higher than Russia (calculated using a percentage difference equation), which is the second highest by medal count- again displaying the USA's clear dominance in the Olympics. Russia and Germany are second and third, respectively, and very close to each other with only 191 medals separating them. The UK comes in fourth, with a pretty significant drop in the medal count; there is nearly a 58% difference between the third and fourth-ranked countries, and nearly a 93% difference between the UK and the USA. From the UK down to Hungary, which is in tenth position, there is a gradual decline, going from 2067 medals down to 1135 medals. The visualisation displays a clear tiered structure for the data, with tier 1 displaying USA as the dominant leader, then tier 2 being Russia and Germany who are pretty even but significantly lower than the USA, and then tier 3 which is UK down to Hungary- which are significantly lower than the top 3, but a steady decline amogst them. 

```python
# For further analysis, create a pivot table to display the distribution of medals, amongst the top 10 countries
award_breakdown = pd.pivot_table(
    df[~df['Medal'].isin(['No Medal'])],
    values='ID',
    index='region',
    columns='Medal',
    aggfunc='count',
    fill_value=0
).sort_values('Gold', ascending=False).head(10)

# Ensure columns are in Gold, Silver, Bronze order
award_order = ['Gold', 'Silver', 'Bronze']
award_breakdown = award_breakdown[[col for col in award_order if col in award_breakdown.columns]]

print('Table 2: Medal breakdown by Country (Top 10):')
award_breakdown
```

```python
# Visualise Table 2 as a vertical stacked bar plot
plt.figure(figsize=(10, 20))
award_breakdown.plot(kind='bar', stacked=True, color=['gold', 'silver', '#CD7F32'])
plt.xlabel('Region')
plt.ylabel('Medal Count')
plt.title('Figure 2: Medal Distribution by Countries (Top 10)')
plt.legend(title='Medal')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('Figure 2: Medal Distribution by Countries (Top 10).png', dpi=300)
plt.show()
```

Figure 2 visualises the total medal counts of the top 10 performing countries, but the data is broken down to show the distribution of the medals data into gold, silver and bronze medals- opening the data up for a deeper analysis of individual countries' performance. As per the previous visualisation, USA is clearly the front-runner dominating the graph with the highest medal counts, but the visualisation of figure 2 clearly shows it also has the highest number of gold and silver medals; the gold medal bar is taller than the total stack of the bottom 7 countries of the top 10; meaning the USA has more gold medals than the total medal count of the 7 countries- clearly displaying its dominance in the Olympic games. Another interesting visualisation is that Italy is ranked above France, despite France having a larger total medal count (larger bar)- this is because Italy has more gold medals than France (its yellow bar is taller). This is because the bar plot is sorted by gold medal count. One final visualisation is that whilst most countries have a relatively balanced ratio of medal distribution; the length of the three coloured bars of each stack are relatively equal. The USA, however, has a gold-heavy ratio- its gold bar is significantly taller than its silver or bronze bars, meaning that they win more on average than any of the other countries. 

```python
# Analysis for most decorated athletes per sport, for the top 10 most decorated athletes
# Simplifyig the code for ease of use
# This code is to filter out the 'No Medal' which we used to fill in the missing values in the 'Medal' column during data cleaning
medals_df = df[~df['Medal'].isin(['No Medal'])]
athlete_medals = medals_df.groupby(['Sport', 'Name']).size().reset_index(name='Medal')

idx = athlete_medals.groupby('Sport')['Medal'].idxmax()

top_athletes = athlete_medals.loc[idx]
print('Table 3: Most Decorated Athlete per Sport (Top 10)')
top_athletes.sort_values('Medal', ascending=False).head(10)
```

Table 3 identifies the most decorated athletes between the 1896 - 2016 olympic games, ranked in descending order their medal counts, along with the sports they participated in. Michael Phelps is clearly the most decorated athlete, with 28 olympic medal wins in swimming. The next most decorated athlete is Larysa Latynina, with 18 medals won in gymnastics. Along with the most decorated athlete in Olympic history, it also identifies the most decorated athlete per sport, within the top 10 of the most decorated athletes; i.e. Michael Phelps is the most decorated swimmer in olympic history whilst Carl Osburn is the seventh most decorated athlete, but the most decorated one for shooting. 

```python
# Analysis for gender participation of athletes throughout the games
gender_breakdown = df.drop_duplicates(subset='ID')
gender_breakdown = gender_breakdown.groupby(df['year_dt'].dt.year)['Sex'].value_counts()
gender_breakdown = pd.DataFrame(gender_breakdown).reset_index()

print('Table 4: Gender Participation of Athletes from 1896 - 2016:')
gender_breakdown
```

```python
# Visualise Table 4 using a line plot
plt.figure(figsize=(15, 7))
plt.plot(gender_breakdown[gender_breakdown['Sex']=='M']['year_dt'],gender_breakdown[gender_breakdown['Sex']=='M']['count'],label='Male')
plt.plot(gender_breakdown[gender_breakdown['Sex']=='F']['year_dt'],gender_breakdown[gender_breakdown['Sex']=='F']['count'],label='Female')
plt.legend()
plt.xlabel('Year')
plt.ylabel('Athlete Numbers')
plt.title('Figure 3: Gender Participation of Athletes from 1896 - 2016')
plt.grid(True, alpha=0.3)
plt.savefig('Figure 3: Gender Participation of Athletes from 1896 - 2016.png', dpi=300)
plt.show()
```

Figure 3 illustrates the gender participation rate of athletes between the 1896 games till the 2016 games; showing a comparison between the male and female participation rate. One clear observation from the visualisation is that female athletes were not permitted to compete in the Olympics until the 1900 Olympics; this is evident due to the fact that the line plot for female participation doesn't start until 1900. The visualisation illustrates a slow but steady narrowing of the gender disparity of athlete participation; from the 1900 till the 1970's; male participation grew rapidly whilst female participation grew very slowly- however from the 1970's till the 2016 games- the rate of feamle participation accelerated, whilst male participation started to plateau- with 2016 being the narrowest gap between the participation of both genders. This signifies that the gender disparity of athlete participation is narrowing significantly. The visualisation also signifies the exponential growth of the Olympic Games over time as well- shown by the rising numbers of athletes participating. From the early 1990's, an oscillating pattern emerges for the athlete participation of both genders, which has continued till 2016- suggesting there may have been a change to the structure of the Olympic Games. 

```python
# Analysis using a pivot table, displaying the number of medals won distributed by region and sport
medals_pivot = pd.pivot_table(
    medals_df,
    values='ID',
    index='NOC',
    columns='Sport',
    aggfunc='count',
    fill_value=0,
    margins=True,
    margins_name='Total'
)

print('Table 5: Medals by Region and Sport:')
print(medals_pivot)
```

Table 5 is a pivot table, which is useful as a comprehensive summary table. It is illustrating the medals won by each individual country, distributed by the sport in which they won the medals. This display is key for identifying whether there are countries which dominate certain sports, or if certain countries focus their resources towards certain sports- and therefore have a higher proportion of medals in those sports. 

```python
# Analysis to display the average ages of athletes participating, throughout the games
avg_age_by_year = df.groupby('Year')['Age'].mean()

# uncomment the lines below to read the breakdown of average age by years
# print("Average age by year:")
# print(avg_age_by_year.round(1))
```

```python
# Visualise the avergae age trend as a line plot
plt.figure(figsize=(12, 6))
avg_age_by_year.plot(kind='line', marker='s', color='blue')
plt.xlabel('Year')
plt.ylabel('Average Age (years)')
plt.title('Figure 4: Average Athlete Age Over Time')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('Figure 4: Average Athlete Age Over Time.png', dpi=300)
plt.show()
```

Figure 4 illustrates the average age trend of athletes participating in the Olympic Games from 1896 till 2016. The visualisation displays three speciifc trends; from the 1900's till the 1940's, despite some volatility- the general trend was that the age of athletes participating was high, with a peak of 29 years in the 1930's. However, since the 1930's till the 1980's- there was a steady and rapid decline in the average age- reaching a low of 23 years old in the 1980 games. From the 1980's till 2016, there has been a trend reverse and the average age rate has started to steadily rise, reaching 26 years old in 2016 games.
